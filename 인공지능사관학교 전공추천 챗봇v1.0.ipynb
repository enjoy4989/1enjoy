```python
from IPython.display import display, clear_output
import ipywidgets as widgets
from datetime import datetime
import time

class IntegratedCareerSystem:
    def __init__(self):
        self.mbti_data = {}
        self.saju_data = {}

    def get_mbti_input(self):
        """MBTI ÏûÖÎ†•ÏùÑ ÏúÑÌïú ÏúÑÏ†Ø ÏÉùÏÑ±"""
        style = {'description_width': '120px'}
        self.mbti_input = widgets.Dropdown(
            options=['INTJ', 'INTP', 'ENTJ', 'ENTP', 'INFJ', 'INFP',
                    'ENFJ', 'ENFP', 'ISTJ', 'ISFJ', 'ESTJ', 'ESFJ',
                    'ISTP', 'ISFP', 'ESTP', 'ESFP'],
            description='MBTI Ïú†Ìòï:',
            style=style
        )
        display(self.mbti_input)

    def get_saju_input(self):
        """ÏÇ¨Ï£º ÏûÖÎ†•ÏùÑ ÏúÑÌïú ÏúÑÏ†Ø ÏÉùÏÑ±"""
        style = {'description_width': '120px'}
        self.name_input = widgets.Text(
            placeholder='ÌôçÍ∏∏Îèô',
            description='Ïù¥Î¶Ñ:',
            style=style
        )
        self.birth_date = widgets.Text(
            placeholder='20000101',
            description='ÏÉùÎÖÑÏõîÏùº:',
            style=style
        )
        self.birth_time = widgets.Text(
            placeholder='0800',
            description='ÏÉùÏãú:',
            style=style
        )
        display(widgets.VBox([self.name_input, self.birth_date, self.birth_time]))

    def validate_inputs(self):
        """ÏûÖÎ†•Í∞í Í≤ÄÏ¶ù"""
        if not self.mbti_input.value:
            print("MBTI Ïú†ÌòïÏùÑ ÏÑ†ÌÉùÌï¥Ï£ºÏÑ∏Ïöî!")
            return False
        if not all([self.name_input.value,
                   self.birth_date.value,
                   self.birth_time.value]):
            print("Î™®Îì† ÏÇ¨Ï£º Ï†ïÎ≥¥Î•º ÏûÖÎ†•Ìï¥Ï£ºÏÑ∏Ïöî!")
            return False
        return True

    def run_analysis(self, btn):
        """Î∂ÑÏÑù Ïã§Ìñâ Ìï∏Îì§Îü¨"""
        clear_output(wait=True)
        if self.validate_inputs():
            self.collect_data()
            self.display_results()

    def collect_data(self):
        """Îç∞Ïù¥ÌÑ∞ ÏàòÏßë"""
        self.mbti_data = {'type': self.mbti_input.value}
        self.saju_data = {
            'name': self.name_input.value,
            'birth_date': self.birth_date.value,
            'birth_time': self.birth_time.value,
            # ÏÇ¨Ï£º Í≥ÑÏÇ∞ Î°úÏßÅ Ï∂îÍ∞Ä
        }

    def display_results(self):
        """Í≤∞Í≥º ÌëúÏãú"""
        print("="*60)
        print(f"üßß {self.saju_data['name']}ÎãòÏùò ÌÜµÌï© Î∂ÑÏÑù Í≤∞Í≥º")
        print("="*60)

        # MBTI Î∂ÑÏÑù Í≤∞Í≥º
        print(f"\nüîÆ MBTI Ïú†Ìòï: {self.mbti_data['type']}")
        print("-"*40)
        print("ÌäπÏßï: Î∂ÑÏÑùÏ†Å ÏÇ¨Í≥†, Ï≤¥Í≥ÑÏ†Å Í≥ÑÌöç ÏàòÎ¶Ω Îä•Î†• Ïö∞Ïàò")

        # ÏÇ¨Ï£º Î∂ÑÏÑù Í≤∞Í≥º
        print(f"\nüìÖ ÏÉùÎÖÑÏõîÏùº: {self.saju_data['birth_date']} {self.saju_data['birth_time']}")
        print("-"*40)
        print("Ïò§Ìñâ Íµ¨ÏÑ±: Êú®(3) ÁÅ´(2) Âúü(1) Èáë(1) Ê∞¥(0)")

        # ÌÜµÌï© Ï∂îÏ≤ú
        print("\nüöÄ Ï∂îÏ≤ú ÏßÑÎ°ú Î∂ÑÏïº:")
        print("-"*40)
        print("1. AI Î™®Îç∏ÎßÅ ÏóîÏßÄÎãàÏñ¥ (ÏûêÏó∞Ïñ¥ Ï≤òÎ¶¨ Ï†ÑÍ≥µ)")
        print("2. Îç∞Ïù¥ÌÑ∞ ÏïÑÌÇ§ÌÖçÌä∏ (ÏãúÏä§ÌÖú ÏÑ§Í≥Ñ Î∂ÑÏïº)")
        print("3. AI ÎπÑÏ¶àÎãàÏä§ Ïª®ÏÑ§ÌÑ¥Ìä∏ (Í∏∞Ïà† Ï†ÑÎûµ Î∂ÑÏïº)")

        print("\nüí° Ï∂îÏ≤ú Ïù¥Ïú†:")
        print("- MBTI Ïú†ÌòïÏùò Î∂ÑÏÑùÎ†•Í≥º ÏÇ¨Ï£ºÏùò Êú® Ïò§ÌñâÏù¥ Í≤∞Ìï©")
        print("- Í∏∞Ïà†Ï†Å Ïó≠ÎüâÍ≥º Ï∞ΩÏùòÏ†Å Î¨∏Ï†ú Ìï¥Í≤∞ Îä•Î†•Ïùò ÏãúÎÑàÏßÄ")

    def run_system(self):
        """ÌÜµÌï© ÏãúÏä§ÌÖú Ïã§Ìñâ"""
        clear_output()
        print("üîÆ MBTI-ÏÇ¨Ï£º ÌÜµÌï© ÏßÑÎ°ú Ï∂îÏ≤ú ÏãúÏä§ÌÖú")

        # MBTI ÏûÖÎ†•
        self.get_mbti_input()

        # ÏÇ¨Ï£º ÏûÖÎ†•
        self.get_saju_input()

        # Ïã§Ìñâ Î≤ÑÌäº
        run_btn = widgets.Button(description="Î∂ÑÏÑù ÏãúÏûë",
                               button_style='success')
        run_btn.on_click(self.run_analysis)
        display(run_btn)

# ÏãúÏä§ÌÖú Ïã§Ìñâ
system = IntegratedCareerSystem()
system.run_system()
```
